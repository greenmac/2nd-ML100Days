# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '../data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train.dtypes.value_counts()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
numeric_columns = []
for col in app_train:
    if app_train[col].dtype == 'float64' or app_train[col].dtype == 'int64' :
        numeric_columns.append(col)
        
# numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])
```

```python
len(numeric_columns)
```

```python
# lambda 是種將運算式 (expression) 重複運用的方式，類似函數 (function)
# 再把只有 2 值 (通常是 0,1) 的欄位去掉 
# 會顯示哪個欄位符合條件 (True 或 False)
condition = list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))

# 只會把 column 是 True 的欄位秀出來
numeric_columns = list(app_train[numeric_columns].columns[condition])
print("Numbers of remain columns" , len(numeric_columns))
```

```python
# 檢視這些欄位的數值範圍
for col in numeric_columns:
    
    print(app_train[col].describe())
    
    fig,axes = plt.subplots(1,1)
    
    # boxes表示箱體，whisker表示觸鬚線
    # medians表示中位數，caps表示最大與最小值界限

    color = dict(boxes='DarkGreen', whiskers='DarkOrange',
              medians='DarkBlue', caps='Red')
    
    # outlier 用 + 表示
    app_train[col].plot(kind='box', ax = axes, subplots = True, color = color, sym ='r+')
    
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)

from statsmodels.distributions.empirical_distribution import ECDF
cdf = ECDF(app_train['AMT_INCOME_TOTAL'])
print(cdf.x)
# 去掉 inf 
print(cdf.x[1:])
print(cdf.y)
# 與 x 相同
print(cdf.y[1:])

plt.plot(list(cdf.x[1:]), cdf.y[1:]/(cdf.y[1:]).max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([(cdf.x[1:]).min(), (cdf.x[1:]).max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.x[1:])), cdf.y[1:]/(cdf.y[1:]).max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
