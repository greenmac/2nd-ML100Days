```python
# 找 CODE_GENDER = F 且 NAME_CONTRACT_TYPE = Revolving loans
condition = np.logical_and(app_train['CODE_GENDER'] == 'F',app_train['NAME_CONTRACT_TYPE'] == 'Revolving loans')
app_train_condition = app_train[condition]
app_train_condition
```

```python
# 算出這群人 AMT_INCOME_TOTAL 的平均
mean = app_train_condition['AMT_INCOME_TOTAL'].mean()
print('平均:', mean)

# 算出這群人 AMT_INCOME_TOTAL 的標準差
std = app_train_condition['AMT_INCOME_TOTAL'].std()
print('標準差:', std)
```

```python
# 計算男生和女生各多少人
male_count = (app_train['CODE_GENDER'] == 'M').sum()
print('Male:', male_count)
female_count = (app_train['CODE_GENDER'] == 'F').sum()
print('Female:', female_count)
N_count = (app_train['CODE_GENDER'] == 'XNA').sum()
print('N_count:', N_count)
```

```python
# 畫出圖
df = pd.DataFrame([male_count, female_count, N_count],
                  index = ['M', 'F', 'XNA'],
                  columns = pd.Index(['Sum'])
)
df.plot(kind='bar')
```
